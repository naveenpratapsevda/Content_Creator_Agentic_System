```python


from youtube_transcript_api import YouTubeTranscriptApi

def extract_transcript(video_url):
    """
    Extracts the raw text from a YouTube video ID or URL.
    """
    try:
        # Extract Video ID from URL if necessary
        video_id = video_url.split("v=")[-1] if "v=" in video_url else video_url
        
        print(f"🔍 Fetching transcript for: {video_id}")
        transcript_data = YouTubeTranscriptApi.get_transcript(video_id)
        
        # Merge timestamps into a single block of text
        raw_text = " ".join([entry['text'] for entry in transcript_data])
        return raw_text
    except Exception as e:
        return f"Error: {str(e)}"

# TEST
# target_video = "https://www.youtube.com/watch?v=VIDEO_ID_HERE"
# raw_transcript = extract_transcript(target_video)
# print(raw_transcript[:500])
```

```python
import modal

# 1. Image: Saare masale (libraries) ek saath
image = (
    modal.Image.debian_slim()
    .pip_install("youtube-transcript-api", "openai")
)

# 2. App: Fortress ka naam
app = modal.App("intelligence-empire", image=image)

# ---------------------------------------------------------
# FUNCTION A: The "Intelligence Scout" (Analysis Agent)
# ---------------------------------------------------------
@app.function(secrets=[modal.Secret.from_name("my-openai-secret")])
def analyze_video_intelligence(transcript: str):
    from openai import OpenAI
    client = OpenAI()
    
    # Is agent ka kaam hai 'Noise' ko 'Signal' mein badalna
    prompt = f"""
    Analyze this transcript and extract the 'Sovereign Alpha':
    1. CATEGORY: Primary niche.
    2. THE ALPHA: Top 3 unique insights that are not common knowledge.
    3. ARBITRAGE POTENTIAL: Why will this go viral in India? (Max 2 sentences).
    
    TRANSCRIPT: {transcript[:5000]}
    """
    
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3 # Accuracy high rakho
    )
    return response.choices[0].message.content

# ---------------------------------------------------------
# FUNCTION B: The "Narrative Architect" (Scripting Agent)
# ---------------------------------------------------------
@app.function(secrets=[modal.Secret.from_name("my-openai-secret")])
def generate_hinglish_script(intelligence_report: str):
    from openai import OpenAI
    client = OpenAI()
    
    # Is agent ka kaam hai data ko 'Entertainment' mein badalna
    prompt = f"""
    Use this Intelligence Report to write a 60-second YouTube Short script.
    - LANGUAGE: Hinglish (Hindi + English).
    - HOOK: Start with a 'Negative Hook' (e.g., 'Aap galat tarike se...').
    - BODY: Use only the 'Alpha' points.
    - CTA: High energy subscribe call.
    
    REPORT: {intelligence_report}
    """
    
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.8 # Creativity thodi zyada rakho hooks ke liye
    )
    return response.choices[0].message.content
```
